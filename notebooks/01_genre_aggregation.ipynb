```python
import os
import pandas as pd
```

```python
directory = 'H:\내 드라이브\datasets\original_crawling'
filenames = os.listdir(directory)
filenames = filenames[:-1]
```

```python
metadata = pd.DataFrame()

for file in filenames:
    temp = pd.read_csv(file, index_col=False)
    temp.drop('Unnamed: 0', axis=1, inplace=True)
    metadata= pd.concat([metadata, temp], ignore_index=True)
    
```

```python
metadata[metadata.duplicated(['Title', 'Artist'], keep=False)]
```

```python
metadata_unique = metadata.drop_duplicates(['Title', 'Artist'])
metadata_unique = metadata_unique.reset_index(drop=True)
metadata_unique['Id'] = metadata_unique.index
metadata_unique = metadata_unique[['Id', 'Title', 'Artist', 'Genre']]
```

```python
metadata_unique.to_csv('H:\\내 드라이브\\datasets\\metadata.csv', index=False)
```
