```python
import pandas as pd
import subprocess
```

### 0. functions

```python
def download_audio(id, video_url, output_path):
    # Command to download audio using yt-dlp

    output_file = f"{output_path}/{id}.wav" #운영체제에 맞게 수정

    command = [
        'yt-dlp',
        '-x',  # Extract audio
        '--audio-format', 'wav',  # Specify audio format
        '-o', output_file,  # Output file path
        video_url,  # URL of the YouTube video
    ]

    try:
        # Execute the command
        subprocess.run(command, check=True)
        print(f"Audio downloaded successfully at: {output_path}")
    except subprocess.CalledProcessError as e:
        print(f"Error downloading audio: {e}")
```

```python
def download_audio_chunks(config, data):
  sample = [[row[0], row[4]] for row in data.values][config.startindex : config.lastindex] #row[0] = id, row[4] = url
  for id, video_url in sample:
    download_audio(id, video_url, config.output_path)
```
