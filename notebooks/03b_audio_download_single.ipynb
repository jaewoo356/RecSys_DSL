```python
import subprocess

def download_audio(video_url, output_path):
    # Command to download audio using yt-dlp
    command = [
        'yt-dlp',
        '-x',  # Extract audio
        '--audio-format', 'mp3',  # Specify audio format
        '-o', output_path,  # Output file path
        video_url,  # URL of the YouTube video
    ]

    try:
        # Execute the command
        subprocess.run(command, check=True)
        print(f"Audio downloaded successfully at: {output_path}")
    except subprocess.CalledProcessError as e:
        print(f"Error downloading audio: {e}")

video_url = "https://www.youtube.com/watch?v=AjzGnoeSdb0"  # 이것만 수정! #

output_path = "/content/drive/MyDrive/DSL/모델링 추천시스템/%(title)s.%(ext)s"  # Save the file with its title as the name
download_audio(video_url, output_path)
```
